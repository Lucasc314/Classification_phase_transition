```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

```python
#Load datasets
X_train = np.load("X_train.npy")
y_train = np.load("y_train.npy")
X_test = np.load("X_test.npy")
y_test = np.load("y_test.npy")
```

For a PCA we need flattend datasets. This will then be reduced from 40 times 40 to 20 components

```python
shape_train = np.shape(X_train)
shape_test = np.shape(X_test)
X_train = X_train.reshape(shape_train[0],shape_train[1]*shape_train[2])
X_test = X_test.reshape(shape_test[0],shape_test[1]*shape_test[2])
X_train.shape
```

Here we can see that the spinconfiguration has been flattened from 40 times 40 to 1600. Let's now begin with the dimensionally reduction of the datasets

```python
L = 20
pca1 = PCA(n_components = L)
pca2 = PCA(n_components = L)
```

```python
pca1.fit(X_train)
X_train_reduced = pca1.transform(X_train)
```

```python
pca1.fit(X_test)
X_test_reduced = pca1.transform(X_test)
```

The componentsize has now been reduced to 20 per spin configuration, which can be seen here:

```python
X_train_reduced.shape
```

Finally, let's export the reduced datasets again

```python
np.save("X_test_reduced.npy", X_test_reduced)
np.save("X_train_reduced.npy", X_train_reduced)
```
